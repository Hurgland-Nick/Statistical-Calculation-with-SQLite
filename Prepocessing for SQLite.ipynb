# Discovery of the dataset



```python
import pandas as pd
import pandas_profiling as pp
import sqlite3

df = pd.read_csv('academy_awards.csv',encoding ='ISO-8859-1')
```

```python
# Overview of the csv.
df.head()
```

```python
# See if we can delete the columns unnamed, see if there's no importante values for every column.
df['Unnamed: 5'].value_counts()
```

```python
# Dimension
df.shape
```

# Filter the data

```python
df['Year'].value_counts()
```

```python
# Part year in two columns, we take the first 4 item
df['Year'] = df['Year'].str[0:4]
```

```python
# type of columns
df['Year'].dtypes
```

```python
# We can see that Year isn't in Integer
# Now we will convert Year in Integer
df['Year'] = df['Year'].astype(int)
df['Year'].dtypes
```

```python
# Keep data where year > 2000
later_than_2000 = df[df['Year'] > 2000]
```

```python
# Show that new variable
later_than_2000.head()
```

```python
# Dimension 
later_than_2000.shape
```

```python
# Actor -- Leadind Role
# Actor -- Supporting Role
# Actress -- Leadind Role
# Actress -- Supporting Role
awards_categories = ['Actor -- Leading Role', 'Actor -- Supporting Role' , 
                     'Actress -- Leading Role' ,'Actress -- Supporting Role']
```

```python
# Filter easily
nominations = later_than_2000[later_than_2000["Category"].isin(awards_categories)]
```

```python
# Dimension
nominations.shape
```

```python
display(nominations)
```

# Clean the Unnamed and Won columns

```python
# Let's create a dictionary to replace "NO" ET "YES" respectively by 0 & 1
replacements = {"NO":0,"YES":1}
```

```python
nominations['Won?'] = nominations['Won?'].map(replacements)
```

```python
nominations.head(30)
```

```python
# Let's create a new colonne named Won without question mark
nominations['Won'] = nominations['Won?']
```

```python
# Show nomination
nominations.head()
```

```python
# Deleted unused columns
# List with columns to delete
drop_cols = ["Won?","Unnamed: 5","Unnamed: 6","Unnamed: 7","Unnamed: 8","Unnamed: 9","Unnamed: 10"]
```

```python
final_nominations = nominations.drop(drop_cols, axis = 1)
final_nominations.head()
```

# Clean Additionnal info columns

```python
# Film {'PERSONNAGE'}

# Biutiful {'Uxbal'}
# The Social Network --> Mark Zuckerberg'
# Clean column
```

```python
additional_info_one = final_nominations['Additional Info'].str.rstrip("'}")
```

```python
additional_info_one.head()
```

```python
# Part the name of film and the actor
additional_info_two = additional_info_one.str.split(" {'", expand = True)
```

```python
additional_info_two.head()
```

```python
movie_names =  additional_info_two[0]
movie_names.head()
```

```python
Characters = additional_info_two[1]
Characters.head()
```

```python
# ADD movie_names & Characters on final_nomination
final_nominations["Movie"] = movie_names
final_nominations["Character"] = Characters
final_nominations.head()
```

```python
# Drop Additional Info column
final_nominations.drop(["Additional Info"], axis = 1, inplace = True)
```

```python
final_nominations.head()
```

# Export to SQLite

```python
# to_sql(name, connexion)
```

```python
final_nominations.dtypes
```

```python
# Let's create a database
connection = sqlite3.connect('nomination.db')
```

```python
final_nominations.to_sql('nomination',connections,  if_exists = 'replace', index = False)
```
